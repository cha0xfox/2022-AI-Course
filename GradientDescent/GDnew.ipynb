```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(w1,w2,x):
    yhat = w1 + w2 * x
    return yhat

def dxW1(w1,w2,x,y):
    yhat = f(w1,w2,x)
    gradient = 2*(yhat - y)
    return gradient

def dxW2(w1,w2,x,y):
    yhat = f(w1,w2,x)
    gradient = 2*x*(yhat - y)
    return gradient

def gradientW1(w1,w2,xs,ys):
    N = len(xs)
    total = 0
    for x,y in zip(xs,ys):
        total += dxW1(w1,w2,x,y)
    gradient = total/N
    return gradient

def gradientW2(w1,w2,xs,ys):
    N = len(ys)
    total = 0
    for x,y in zip(xs,ys):
        total += dxW2(w1,w2,x,y)
    gradient = total/N
    return gradient

def gradDesc(xs,ys,lr,N):
    w1 = np.random.uniform(0,1,1)
    w2 = np.random.uniform(0,1,1)
    print(w1)
    w1path = []
    w2path = []
    
    for i in range(N):
        w1 = w1 - lr * gradientW1(w1,w2,xs,ys)
        w2 = w2 - lr * gradientW2(w1,w2,xs,ys)
        w1path.append(w1)
        w2path.append(w2)

    return (w1,w2),w1path,w2path
```

```python
xs = [1,2,3,4,5,6,7]
ys = [2,4,6,8,10,12,14]
(w1,w2),w1path,w2path = gradDesc(xs,ys,lr=0.01,N=400)
print(w1,w2)
```

```python
plt.plot(xs,ys)
plt.plot(w1path,w2path)
```
