```python
import numpy as np
import plotly as px
import pandas as pd
```

## Norm

```python
def norm(x1,x2):
    return np.random.random() * (x2-x1) + x1

def normdata(n):
    x = []
    for i in range(n):
        x.append(norm(0,1))
    return x
```

```python
df = pd.DataFrame(dict(value = normdata(10)))
fig = px.express.line(df)
fig.show()
print(df)
```

## Gauss

```python
def gauss(m=0, d=1):
    '''
    m - moment
    d - deviation
    '''
    n = 12
    x = 0
    for i in range (n):
         x = x + np.random.random()
    return d * (12/n)**(1/2) * (x - n/2) + m

def gaussdata(n):
    x = []
    for i in range(n):
        x.append(gauss())
    return x
```

```python
df = pd.DataFrame(dict(value = gaussdata(10)))
fig = px.express.line(df)
fig.show()
print(df)
```

## Two Peaks

```python
def twopeaksdata(N,x1=-2,y1=2,x2=-4,y2=4):
    '''
    N - iterations
    x1,y1,x2,y2 - gauss params
    '''
    x = []
    for i in range(N):
        r = norm(0,1)
        if r > 0.6:
            x.append(gauss(x1,x2))
        else:
            x.append(gauss(y1,y2))
    return x
```

```python
df = pd.DataFrame(dict(value = twopeaksdata(10)))
fig = px.express.line(df)
fig.show()
print(df)
```

## Parzen-Rosenblatt

```python
def gausskernel(r):
     return ((2 * np.pi)**(-1/2)) * np.exp(-1/2 * (r**2))
    
def parzenrozenblatt(x, data, h):
    rez = 0
    for i in data:
        rez = rez + gausskernel((x - i)/h)
    return rez / (len(data)*h)
```

## Leave one out
