# Лабораторная №1
### Градиентный спуск

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def func(x):
    return 2 * x**2 - 4 * x #2*x^2-4*x
    
def grad(x):
    return (4*x) - 4

def gradDesc(lr = 0.15,start = 10.0,N = 15):
    xp = np.empty(N)
    xp[0] = start

    for i in range(1,N):
        d = grad(xp[i-1])
        xp[i] = xp[i-1] - (d * lr)
    return xp
```

```python
xp = np.linspace(-10,10,100)
yp = func(xp)

plt.plot(xp,yp)

xpath = gradDesc(lr = 0.34, N = 15)

print(xpath)

plt.plot(xpath,func(xpath),'-o')
```
